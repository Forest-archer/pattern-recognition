```python
import os
data=[]
for file in os.listdir("r8"):
    data.append(file)
for name in data:
    f=open("r8/"+name,'r')
    points=f.read().split("\n")
    f.close()
    print(name)
    for i in range(0,len(points)-1):
        cut=points[i].split(" ")
        
        if i == 0:
            f=open("C1_points/"+name,'w')
        else:
            f=open("C1_points/"+name,'a')
        #print(i)
        #print(cut)
        f.write(cut[0]+" "+cut[1]+" "+cut[2]+"\n")
        f.close()
        if i == 0:
            fd=open("C1_label/"+name,'w')
        else:
            fd=open("C1_label/"+name,'a')
        fd.write(str(int(float(cut[3])))+"\n")
        fd.close()
```
