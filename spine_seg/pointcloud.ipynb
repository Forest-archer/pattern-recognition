```python
from pyntcloud import PyntCloud
import pyvista as pv
import os

for file in os.listdir("SPAISTL"):
    name=file.split(".")
    anky = PyntCloud.from_file("SPAISTL/"+file)
    anky.points
    anky_cloud = anky.get_sample("mesh_random", n=48000, rgb=False, normals=True, as_PyntCloud=True)
    anky_cloud.to_file("SPAIPTS/"+name[0]+".pts",sep=" ",header=0,index=0)
```
